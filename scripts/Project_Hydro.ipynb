```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Your existing data loading and preprocessing steps
# Strip whitespace from station names
df['stacja'] = df['stacja'].str.strip()

# Convert the datetime column to datetime type
df['stan_wody_data_pomiaru'] = pd.to_datetime(df['stan_wody_data_pomiaru'])

# Define the list of rivers to filter (added 6 rivers as requested)
rivers = ['Żukowo', 'Władysławowo', 'Tujsk', 'Ustka', 'Tolkmicko', 'Hel']

# Define your date range
start_date = '2024-10-04 10:10:00'
end_date = '2024-10-05 06:10:00'

# Create a figure with subplots (2 rows, 3 columns, to accommodate 6 rivers)
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Loop through each river and create a separate plot for each one
for i, river in enumerate(rivers):
    # Filter the DataFrame for the selected river and date range
    filtered_df = df[
        (df['stacja'] == river) &
        (df['stan_wody_data_pomiaru'] >= start_date) &
        (df['stan_wody_data_pomiaru'] <= end_date)
    ]
    
    # Calculate the moving average safely
    filtered_df['moving_avg'] = filtered_df['stan_wody'].rolling(window=3).mean()

    # Plot the water level and moving average
    sns.lineplot(data=filtered_df, x='stan_wody_data_pomiaru', y='stan_wody', color='red', ax=axes[i], label='Water Level')
    sns.lineplot(data=filtered_df, x='stan_wody_data_pomiaru', y='moving_avg', color='red', linestyle='--', ax=axes[i], label='Moving Average')

    # Set the title for each subplot
    axes[i].set_title(f'Water Level Trends of {river}')

    # Rotate x-axis labels for better readability
    axes[i].tick_params(axis='x', rotation=45)
    
    # Set axis labels
    axes[i].set_xlabel('Date')
    axes[i].set_ylabel('Water level')

# Remove individual legends from all subplots
for ax in axes:
    ax.legend().remove()

handles, labels = axes[1].get_legend_handles_labels()
axes[1].legend(handles, ['Water Level', 'Moving Average'], loc='upper center', bbox_to_anchor=(0.5, 1.2), ncol=2, title='Legend')

# Adjust layout to avoid overlapping
plt.tight_layout()
plt.savefig('D:\\end-to-end_data_engineer_project\\water_level_trends_selected_high_discharge_rivers.png', dpi=300)
plt.show()
```
